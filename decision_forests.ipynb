```python
# Load TensorFlow Decision Forests
import tensorflow_decision_forests as tfdf
```

```python
# Load training  dataset using pandas
import pandas as pd
```

```python
file_path =  "/content/drive/MyDrive/aprofundamento/pinguins_arvore_de_decisao/penguins.csv"
train_df = pd.read_csv(file_path)
train_df.head()
```

```python
import numpy as np
```

```python
def split_dataset(dataset, test_ratio=0.30):
  test_indices = np.random.rand(len(dataset)) < test_ratio
  return dataset[~test_indices], dataset[test_indices]

train_ds_pd, valid_ds_pd = split_dataset(train_df)
print("{} exemplos em treinamento, {} exemplos em teste.".format(len(train_ds_pd), len(valid_ds_pd)))
```

```python
# Convert the pandas dataframe into a Tensorflow dataset
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_ds_pd, label='species')
```

```python
# Train the model
model = tfdf.keras.RandomForestModel()
model.fit(train_ds)
```

## Evaluating Model

```python
# Load the testing dataset
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(valid_ds_pd, label="species")
```

```python
# Evaluate the model
model.compile(metrics=["accuracy"])
print(model.evaluate(test_ds))
```

## Interpretando o modelo

```python
tfdf.model_plotter.plot_model_in_colab(model, tree_idx=0)
```
